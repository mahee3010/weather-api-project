```python
import os
import requests

url = "https://weather-api167.p.rapidapi.com/api/weather/forecast"

querystring = {"place":"London,GB","cnt":"3","units":"standard","type":"three_hour","mode":"json","lang":"en"}

headers = {
	"x-rapidapi-key": os.getenv("RAPIDAPI_KEY"),
	"x-rapidapi-host": "weather-api167.p.rapidapi.com",
	"Accept": "application/json"
}

response = requests.get(url, headers=headers, params=querystring)

```

```python
response.json()['list']
```

```python
import pandas as pd
```

```python
df=pd.json_normalize(response.json()['list'])
```

```python
df
```

```python
df["weather_main"]=df["weather"].apply(lambda x:x[0]["main"])
df["weather_description"]=df["weather"].apply(lambda x:x[0]["description"])
df.drop(columns=["weather"],inplace=True)
```

```python
df
```

```python
columns_to_keep = ["dt","main.temprature","main.temprature_feels_like","main.temprature_max","main.temprature_min","main.pressure","main.humidity","wind.speed","wind.degrees","weather_main","weather_description","clouds.cloudiness","sys.part_of_day","visibility_distance"]
```

```python
df=df[columns_to_keep]
```

```python
df.to_csv('weather.csv')
```
